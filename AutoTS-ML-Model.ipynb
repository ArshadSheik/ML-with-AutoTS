# AutoTS

AutoTS means Automatic Time Series, it is a machine learning library in Python used for the task of time series forecasting. It can be used for the task of time series forecasting. Some of the popular features of the AutoTS library in Python are:

1. It can be used to find the best time series forecasting model which depends on the type of data you are using.

2. It can handle both the univariate and multivariate time series.

3. It can also handle messy data by removing and filling the NaN values and it can also handle the outliers.

4. The models provided by this Python library for can be used for deployment also.

## Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("AAPL 1.csv")
```

## Preparing data

```python
data = data[["Date", "Close"]]
data["Date"] = pd.to_datetime(data.Date)
data["Close"].plot(figsize=(12, 8), title="Apple Stock Prices", fontsize=20, label="Close Price")
plt.legend()
plt.grid()
plt.show()
```

## Implementing AutoTS

```python
from autots import AutoTS

model = AutoTS(forecast_length=10, frequency='infer',
            ensemble='simple', drop_data_older_than_periods=200, max_generations=10)

model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
```

```python
prediction = model.predict()
forecast = prediction.forecast
print("Stock Price Prediction of Apple")
print(forecast)
```
